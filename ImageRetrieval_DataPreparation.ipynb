```python
import numpy as np
from tensorflow.keras.preprocessing import image as kimage
from tqdm import tqdm 

images = []
lists = [i for i in range(9908)]

for index in tqdm(lists):
    img = kimage.load_img('images_mr/%d.jpg' % (index), target_size=(86, 128))
    img_np = kimage.img_to_array(img) 
    images.append(img_np)
    
images_np = np.array(images)
print(images_np.shape)

# save dataset
np.save('images_mr1.npy', images_np)
```

```python
print(images_np.shape)
print(type(images_np[0,0,0,0]))
```

```python
import numpy as np
from tensorflow.keras.preprocessing import image as kimage
from tqdm import tqdm 

images = []
lists = [i for i in range(9908)]
for index in tqdm(lists):
    img = kimage.load_img('images_mr/%d.jpg' % (index), target_size=(86, 128))    
    img_np = kimage.img_to_array(img)
    images.append(img_np)
    
# convert to np.array
images_np = np.array(images)
print(images_np.shape)

# reduce memory used
images_np = images_np.astype(np.uint8)
np.save('images_mr2.npy', images_np)
```

```python
import numpy as np

data = np.load('images_mr2.npy', allow_pickle=True)
print(data.shape)

import matplotlib
import matplotlib.pyplot as plt

plt.figure(figsize=(4,4))
plt.imshow(data[0]/255.0)
```
